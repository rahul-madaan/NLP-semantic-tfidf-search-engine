##TF-IDF processing


```python
'''
using the tfidf vectorizer of sklearn library for 
efficient calculations and use of disk and RAM
'''

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import nltk

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
# noramlization case lowering
def lower_case(lines):
  lines = lines.lower()
  return lines

# removing punctuation ie -> !"#$%&'()*+,-./:;<=>?@[\]^_`{|}~
import string
import re
def remove_punct(text):
    text  = "".join([char for char in text if char not in string.punctuation])
    text = re.sub('[0-9]+', '', text)
    return text

def tokenization(text):
  text = text.strip()
  text = re.split('\W+', text)
  return text

def remove_singles(text):
#removing all 1 letter words like 's', 'A' etc
  for i in text:
    if len(i) == 1:
      text.remove(i)
  return text

stopwords = nltk.corpus.stopwords.words('english')
def remove_stopwords(text):
    text = [word for word in text if word not in stopwords]
    return text

stopwords = nltk.corpus.stopwords.words('english')
def remove_stopwords(text):
    text = [word for word in text if word not in stopwords]
    return text

def stemming(text):
    text = [nltk.PorterStemmer().stem(word) for word in text]
    return text

def lemmatizer(text):
    text = [nltk.WordNetLemmatizer().lemmatize(word) for word in text]
    return text

```

```python
# storing all the sentences in a list for easy traversal
from itertools import chain
flatten_list1 = list(chain.from_iterable(news_df.values.tolist()))
```

```python
# data preprocessing
for i in range(len(flatten_list1)):
  flatten_list1[i] = lower_case(flatten_list1[i])
  flatten_list1[i] = remove_punct(flatten_list1[i])
  flatten_list1[i] = tokenization(flatten_list1[i])
  flatten_list1[i] = remove_singles(flatten_list1[i])
  flatten_list1[i] = stemming(flatten_list1[i])
  flatten_list1[i] = lemmatizer(flatten_list1[i])
  flatten_list1[i] = remove_stopwords(flatten_list1[i])
  flatten_list1[i] = " ".join(flatten_list1[i])

print(flatten_list1)
```

```python
vectorizer = TfidfVectorizer() #initializing the TF-IDF vercorizer form sklearn lib
X = vectorizer.fit_transform(flatten_list1)
print(X.shape) #(no of headlines, no. of unique words)
```

```python
#query selection
query = queries[0]

#query processing, same as data pre processing
query = lower_case(query)
query = remove_punct(query)
query = tokenization(query)
query = remove_singles(query)
query = stemming(query)
query = lemmatizer(query)
query = remove_stopwords(query)
query = " ".join(query)
result_tfidf_sentences = []
result_tfidf_scores = []
#finding vector of query
query_vec = vectorizer.transform([query]) # Ip -- (n_docs,x), Op -- (n_docs,n_Feats)
results = cosine_similarity(X,query_vec).reshape((-1,)) # Op -- (n_docs,1) -- Cosine Sim with each doc
#comparing and finding the nearest 10 results
#comparing the cosine scores
for i in results.argsort()[-10:][::-1]:
  result_tfidf_sentences.append(news_df.iloc[i,0])
  result_tfidf_scores.append(results[i])

print(result_tfidf_sentences)
print(result_tfidf_scores)
```

```python
result_dict = {} #final dictioanry

for i in range(len(top_10_values_semantic)): #instering the semantic results into the dictionary in begining
  result_dict[top_10_values_semantic[i]] = float(str(top_10_scores_semantic[i])[9:15]) 

for i in range(len(result_tfidf_sentences)): #instering the tf-idf results
  if result_tfidf_sentences[i] not in result_dict.keys():#if the sentences DO NOT overlap in result sets of tf-idf and semantic models.
    result_dict[result_tfidf_sentences[i]] = result_tfidf_scores[i]
  else: # if the sentences DO overlap in result sets of tf-idf and semantic models.
    if result_dict[result_tfidf_sentences[i]] <= 0.6: # +40%
      result_dict[result_tfidf_sentences[i]] = result_dict[result_tfidf_sentences[i]] + (0.4*result_dict[result_tfidf_sentences[i]])
    elif result_dict[result_tfidf_sentences[i]] <= 0.7: # +30%
      result_dict[result_tfidf_sentences[i]] = result_dict[result_tfidf_sentences[i]] + (0.3*result_dict[result_tfidf_sentences[i]])
    elif result_dict[result_tfidf_sentences[i]] <= 0.8: # +20%
      result_dict[result_tfidf_sentences[i]] = result_dict[result_tfidf_sentences[i]] + (0.2*result_dict[result_tfidf_sentences[i]])
    elif result_dict[result_tfidf_sentences[i]] <= 0.9: # +10%
      result_dict[result_tfidf_sentences[i]] = result_dict[result_tfidf_sentences[i]] + (0.1*result_dict[result_tfidf_sentences[i]])
    else: #max score can be between [0.0,1.0]
      result_dict[result_tfidf_sentences[i]]=1.0
#sorting the dictioanry by values
result_dict = {k: v for k, v in sorted(result_dict.items(), key=lambda item: item[1],reverse=True)}

for i in result_dict :
    print(i, result_dict[i])
    print("")
```
